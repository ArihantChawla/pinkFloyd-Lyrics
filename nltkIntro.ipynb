```python
import nltk
from nltk import FreqDist
import csv
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from collections import OrderedDict
```

```python
with open ("lyricsAll.txt", "r") as myfile:
    text=myfile.read().replace('\n', '')
```

```python
tokenizer = RegexpTokenizer(r'\w+')
nltext=nltk.Text(tokenizer.tokenize(text))
```

```python
print (nltext)
```

```python
nltext[1024:1062]
```

```python
nltext.collocations()
```

```python
filtered_words=[word for word in nltext if word.lower() not in stopwords.words('english') ]
```

```python
print (filtered_words)
```

```python
commonText=nltk.Text(filtered_words)
```

```python
commonText
```

```python
freq=FreqDist(commonText)
```

```python
visualize=(freq.most_common(32))
#Since first 2 entries are 're' and 'll' and make no sense , I am manually removing it.
visualize=visualize[2:]
```

```python
dV=dict(visualize)
```

```python
finalData=OrderedDict(sorted(dV.items(), key=lambda t: t[1],reverse=True))
```

```python
print (finalData)
```

```python
writer = csv.writer(open('visual.csv', 'a'))
for key, value in finalData.items():
   writer.writerow([key, value])
```
